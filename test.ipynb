```python
def generate_surrounding_coordinates(center_lat, center_lon, step=0.01, range_deg=0.1):
    """
    Generate a grid of lat/lon values around a center point.

    - step: degree increment (~0.01° ≈ 1.1 km)
    - range_deg: how far to go from center in degrees (e.g. 0.1°)
    """
    lat_range = [center_lat + i * step for i in range(int(-range_deg/step), int(range_deg/step)+1)]
    lon_range = [center_lon + i * step for i in range(int(-range_deg/step), int(range_deg/step)+1)]

    grid_coords = [(lat, lon) for lat in lat_range for lon in lon_range]
    return grid_coords

# Example
center_lat = 44.4473
center_lon = -74.9302
nearby_coords = generate_surrounding_coordinates(center_lat, center_lon, step=0.01, range_deg=0.05)

for coord in nearby_coords[:]:  # print only first 10
    print(coord)
```
